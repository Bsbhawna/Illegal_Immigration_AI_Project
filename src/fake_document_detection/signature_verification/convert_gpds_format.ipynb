```python
import os
import shutil

SOURCE_DIR = "signature_verification/download/test"
DEST_DIR = "signature_verification/data"

os.makedirs(DEST_DIR, exist_ok=True)

for user_id in os.listdir(SOURCE_DIR):
    user_path = os.path.join(SOURCE_DIR, user_id)
    if not os.path.isdir(user_path): continue

    genuine_path = os.path.join(user_path, "genuine")
    forge_path = os.path.join(user_path, "forge")
    new_user_folder = os.path.join(DEST_DIR, f"full_org_{user_id.zfill(3)}")
    os.makedirs(new_user_folder, exist_ok=True)

    # Copy genuine signatures
    for i, file in enumerate(sorted(os.listdir(genuine_path))):
        src = os.path.join(genuine_path, file)
        dst = os.path.join(new_user_folder, f"org_{user_id.zfill(3)}_{str(i+1).zfill(2)}.png")
        shutil.copy(src, dst)

    # Copy forged signatures
    for i, file in enumerate(sorted(os.listdir(forge_path))):
        src = os.path.join(forge_path, file)
        dst = os.path.join(new_user_folder, f"for_{user_id.zfill(3)}_{str(i+1).zfill(2)}.png")
        shutil.copy(src, dst)

print("✅ Folder restructuring complete.")
```

```python
from pathlib import Path

base_dir = Path("signature_verification/data")
for user_folder in base_dir.iterdir():
    if user_folder.is_dir():
        genuine = list(user_folder.glob("org_*.png"))
        forged = list(user_folder.glob("for_*.png"))
        print(f"{user_folder.name} → Genuine: {len(genuine)}, Forged: {len(forged)}")
```
